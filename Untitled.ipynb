```julia
# == Model parameters == #
r       = 0.05
bet     = 1 / (1 + r)
T       = 50
c_bar   = 1.5
sigma   = 0.15
mu      = 2
q       = 1e4
m1      = T * (mu / (T/2)^2)
m2      = - (mu / (T/2)^2)

# == Formulate as an LQ problem == #
Q = 1.0
R = zeros(4, 4)
Rf = zeros(4, 4); Rf[1, 1] = q
A = [1 + r -c_bar m1 m2; 
     0     1      0  0;
     0     1      1  0;
     0     1      2  1]
B = [-1.0; 0.0; 0.0; 0.0]
C = [sigma; 0.0; 0.0; 0.0]

# == Compute solutions and simulate == #
lq = LQ(Q, R, A, B, C; bet = bet, capT = T, rf = Rf)
x0 = [0.0; 1.0; 0.0; 0.0]
xp, up, wp = compute_sequence(lq, x0)

# == Convert results back to assets, consumption and income == #
ap = vec(xp[1, 1:end])                                 # Assets
c = vec(up + c_bar)                                     # Consumption
time = 1:T
income = vec(wp[1, 2:end]) + m1 * time + m2 * time.^2  # Income

# == Plot results == #
plot(Vector[income, ap, c, zeros(T + 1)], lab = ["non-financial income" "assets" "consumption" ""], 
    color = [:orange :blue :green :black], width = 3, xaxis = ("Time"), layout = (2,1),
    bottom_margin = 20mm, size = (900, 900), show = true);
```
